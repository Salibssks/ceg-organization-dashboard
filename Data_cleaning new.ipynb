```python
df['org_clean'] = df['Organisation']
```

```python
df['org_clean'] = df['org_clean'].str.lower()
```

```python
df['org_clean'] = df['org_clean'].str.replace(r'[^a-z0-9\s]', '', regex=True)
```

```python
df['org_clean'] = df['org_clean'].str.strip()
df['org_clean'] = df['org_clean'].str.replace(r'\s+', ' ', regex=True)
```

```python
df[['Organisation', 'org_clean']].head(600)
```

```python
noise_words = [
    "department", "dept", "office", "division", "unit", "cell",
    "circle", "region", "regional", "ro",
    "chief", "engineer", "executive", "superintending",
    "government", "of", "the", "and"
]
```

```python
df['org_clean'] = df['org_clean'].apply(
    lambda x: ' '.join([word for word in str(x).split() if word not in noise_words]) if pd.notnull(x) else x
)
```

```python
noise_words = [word.lower() for word in noise_words]
```

```python
df['org_clean'] = df['org_clean'].apply(
    lambda x: ' '.join(
        [word for word in str(x).lower().split() if word not in noise_words]
    ) if pd.notnull(x) else x
)
```

```python
df[['Organisation', 'org_clean']].head(30)
```

```python
org_map = {
    "nhai": "NHAI",
    "national highways authority": "NHAI",
    
    "airports authority": "AAI",
    "aai": "AAI",
    
    "indian railways": "Indian Railways",
    "railway": "Indian Railways",
    
    "pwd": "PWD",
    
    "municipal": "Municipal Corporation",
    "municipal corporation": "Municipal Corporation"
}
```

```python
def map_org(x):
    if pd.isnull(x):
        return x
    
    for key in org_map:
        if key in x:
            return org_map[key]
    
    return x
```

```python
df['org_final'] = df['org_clean'].apply(map_org)
```

```python
df[['Organisation', 'org_clean', 'org_final']].head(50)
```

```python
df['org_search'] = df['org_clean'].fillna('') + " " + df['org_final'].fillna('')
```

```python
df['org_search'] = df['org_search'].str.lower()
df['org_search'] = df['org_search'].str.replace(r'\s+', ' ', regex=True)
df['org_search'] = df['org_search'].str.strip()
```
